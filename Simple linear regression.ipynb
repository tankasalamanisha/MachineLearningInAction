## Difference between percentile quntile and quartile

0 quartile = 0 quantile = 0 percentile <br>

1 quartile = 0.25 quantile = 25 percentile <br>

2 quartile = .5 quantile = 50 percentile (median) <br>

3 quartile = .75 quantile = 75 percentile <br>

4 quartile = 1 quantile = 100 percentile

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating a dummy dataset
np.random.seed(10)
data = np.random.rand(60,1)
target = 3+3*data + np.random.rand(60,1)
```

```python
data
```

```python
target
```

```python
# scatter plot

plt.scatter(x=data,
            y=target,
            s=10)
plt.xlabel("dummy data")
plt.ylabel("dummy target")
plt.show()
```

## Model

```python
from sklearn.linear_model import LinearRegression
```

```python
help(LinearRegression)
```

```python
lr= LinearRegression()
```

```python
lr.fit(data,target)  # training
```

```python
# predicting on the fit-regressor
pred = lr.predict(data)
```

```python
plt.plot(data,pred,color='black',marker='o',)
```
